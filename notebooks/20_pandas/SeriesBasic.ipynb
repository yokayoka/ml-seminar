```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
# numpy.arrayに似たSeriesをつくります。違いは、データにラベルがあるところ

obj = Series([3, 6, 9, 12])
obj
```

```python
# valuesで、値が返ってくる
obj.values
```

```python
# indexを表示
obj.index
```

```python
# index付きのデータを作る
# 第二次世界大戦の死傷者

ww2_cas = Series([8700000, 4300000, 3000000, 2100000, 400000],index=['USSR', 'Germany', 'China', 'Japan', 'USA'])
ww2_cas
```

```python
# 文字列のindexでアクセスできる。
ww2_cas['USA']
```

```python
# 400万人以上の死傷者を出したのは？

ww2_cas[ww2_cas>4000000]
```

```python
# 普通の辞書のように扱えます。
# USSR があるか？

'USSR' in ww2_cas
```

```python
# 辞書型に変換できます。

ww2_dict = ww2_cas.to_dict()
ww2_dict
```

```python
# 辞書をもとに、Seriesを作ることができます。

WW2_Series = Series(ww2_dict)
```

```python
WW2_Series
```

```python
# indexを明示的に与えることができます。たとえば・・・

countries = ['China', 'Germany', 'Japan', 'USA', 'USSR', 'Argentina']
```

```python
# 別のSeriesを作ります

obj2 = Series(ww2_dict, index=countries)
```

```python
obj2
```

```python
# nullデータがあるかどうかを確認できます

pd.isnull(obj2)
```

```python
# 逆のこともできます。

pd.notnull(obj2)
```

```python
# もとのデータに戻りましょう。

WW2_Series
```

```python
obj2
```

```python
# 両方のデータを足すと、pandasが自動的に、indexでまとめてくれます。

WW2_Series + obj2
```

```python
# Seriesに名前を付けられます。

obj2.name = '第二次世界大戦の死傷者'
```

```python
obj2
```

```python
# indexに名前を付けることも可能

obj2.index.name = 'Countries'
```
